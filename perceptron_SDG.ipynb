### Creating a Perceptron using the **Hinge loss**

```python
from sklearn.datasets import make_classification
import numpy as np
```

```python
X,y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, random_state=41,
                          n_classes=2, n_clusters_per_class=1, hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def Activation_function(z):
    return 1 if z>0 else 0
```

```python
def Percepteron(X,y):
    
    w1=w2=b=1
    lr = 0.01

    
    for i in range(1000):
        # applying gradient descent to find w1 ,w2,b
        for j in range(X.shape[0]):
            z = w1*X[j][0]  + w2*X[j][1] + b

            if z*y[j] <= 0:
                w1 = w1 + lr * y[j] * X[j][0]
                w2 = w2 + lr * y[j] * X[j][1]
                b = b + lr * y[j]

    return w1,w2,b
```

```python
w1,w2,b = Percepteron(X,y)
```

```python
m = -(w1/w2)
c = -(b/w2)
```

```python
X_input = np.linspace(-3,3,100)
Y_input = m*X_input +c
```

```python
plt . figure(figsize=(10,6))

plt.plot(X_input,Y_input,color='red' , linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y , cmap='winter' , s=100)
plt.ylim(-3,2)
```
