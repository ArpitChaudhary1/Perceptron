## Creating a perceptron using the **perceptron trick** to find weights and bias

```python
from sklearn.datasets import make_classification
import numpy as np
```

```python
X,y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, random_state=41,
                          n_classes=2, n_clusters_per_class=1, hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def Activation_function(z):
    return 1 if z>0 else 0
```

### Created a fuction Perceptron 

```python
def Percepteron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.01

    for i in range(1000):
        j = np.random.randint(0,100)
        y_predict = Activation_function(np.dot(X[j],weights))
        weights = weights + lr*(y[j] - y_predict)*X[j]

    return weights[0], weights[1:]
```

```python
intercept_ , coef_ = Percepteron(X,y)
```

intercept_ is the bias ans coef_ is the weights

```python
print(intercept_)
print(coef_)
```

Here we are using the general equation of line **Ax + By + c = 0**. <br>
So to find the slope and intercept of the line I am going to use the formulae: <br>

1. slope(m) = -(A/B)
2. intercept (c) = -(C/B)

```python
m = -(coef_[0]/coef_[1])
c = -(intercept_/coef_[1])

X_input = np.linspace(-3,3,100)       # some points of to draw the line     linespace create a numpy array 
Y_input = m*X_input + c
```

```python
plt . figure(figsize=(10,6))

plt.plot(X_input,Y_input,color='red' , linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y , cmap='winter' , s=100)
plt.ylim(-3,2)
```
